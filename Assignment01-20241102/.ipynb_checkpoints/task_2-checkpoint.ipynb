```python
from skimage import io, color, img_as_float
from skimage.transform import integral_image
import matplotlib.pyplot as plt
```

```python
image_path = 'visual_attention_ds.png' 
image_rgba = io.imread(image_path)
```

```python
image_rgb = color.rgba2rgb(image_rgba)
```

```python
image_gray = color.rgb2gray(image_rgb)
```

```python
integral_img = integral_image(image_gray)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image_rgb)

plt.subplot(1, 2, 2)
plt.title("GrayScale Integral Image")
plt.imshow(integral_img, cmap="gray")
plt.show()
```

```python
def compute_contrast_map(integral_img, center_size, surround_size):
    center_half = center_size // 2 # Calculating half
    surround_half = surround_size // 2

    contrast_map = np.zeros_like(integral_img)

    for y in range(surround_half, integral_img.shape[0] - surround_half - 1):
        for x in range(surround_half, integral_img.shape[1] - surround_half - 1):
            # Center window coordinates
            top_c, bottom_c = y - center_half, y + center_half + 1
            left_c, right_c = x - center_half, x + center_half + 1

            # Surround window coordinates 
            top_s, bottom_s = y - surround_half, y + surround_half + 1
            left_s, right_s = x - surround_half, x + surround_half + 1

            center_sum = (integral_img[bottom_c, right_c] - integral_img[top_c, right_c] - integral_img[bottom_c, left_c] + integral_img[top_c, left_c])
            surround_sum = (integral_img[bottom_s, right_s] - integral_img[top_s, right_s] - integral_img[bottom_s, left_s] + integral_img[top_s, left_s])

            # Contrast value
            contrast_map[y, x] = surround_sum - center_sum

    contrast_map = (contrast_map - np.min(contrast_map)) / (np.max(contrast_map) - np.min(contrast_map))
    return contrast_map
        
```

```python
window_sizes = [(11, 21), (3, 7), (31, 51)]
fig, axes = plt.subplots(1, len(window_sizes), figsize=(15, 5))

for i, (center_size, surround_size) in enumerate(window_sizes):
    contrast_map = compute_contrast_map(integral_img, center_size, surround_size)
    axes[i].imshow(contrast_map, cmap='gray')
    axes[i].set_title(f"Center: {center_size}x{center_size}, Surround: {surround_size}x{surround_size}")
    axes[i].axis('off')

plt.show()
```
