### Grayscale Histogram Calculation

1. **Convert Images to Grayscale**:
   - For each image, convert it to grayscale using the averaging method. The grayscale value for each pixel can be calculated as:
     \[
     \text{Grayscale} = \frac{R + G + B}{3}
     \]
   - Where \( R \), \( G \), and \( B \) are the pixel values of the red, green, and blue channels, respectively.

2. **Define Histogram Bins**:
   - Create a fixed set of 51 bins that cover the pixel value range from 0 to 255. Each bin should have a length of 5. You can define the bins as follows:
     ```python
     bins = np.arange(0, 256, 5)  # This will create 51 bins
     ```

3. **Calculate Histograms**:
   - For each grayscale image, calculate the histogram using the predefined bins. 
   - Do not use the automatically chosen bins by `numpy.histogram`, instead utilize the vector for bins that you created.

4. **Store Histograms**:
   - Store the computed histograms in a dictionary or list for later use. This will facilitate easy access when comparing histograms during the test phase.

5. **Considerations**:
   - Think about why it is necessary to use fixed bins rather than automatically chosen bins. 
     - **Hint**: If bins are chosen automatically, the histograms may not be comparable because they could be based on different ranges or intervals depending on the image characteristics.


````python
import numpy as np
import matplotlib.pyplot as plt

batch_1_g = {1: [], 4: [], 8: []}

for value in [1, 4, 8]:
    for image in batch_1_f[value]:  # Access images directly with the value
        # Reshape the image to (32, 32, 3) if it's not already in that shape
        image_c = image.reshape(3, 32, 32).transpose(1, 2, 0)
        
        # Compute the grayscale image
        image_g = np.sum(image_c, axis=2) / 3  # Average of the R, G, B channels
        image_g = image_g.astype(np.uint8)  # Ensure the image is in uint8 format

        # Print the shapes for debugging
        print("RGB image shape:", image_c.shape)
        print("Grayscale image shape:", image_g.shape)

        # Display the grayscale image
        plt.imshow(image_g, cmap='gray')  # Use cmap='gray' to display it in grayscale
        plt.axis('off')  # Hide the axes for better visualization
        plt.show()  # Show the image

        # Display the color image
        plt.imshow(image_c)  # Display the color image
        plt.axis('off')  # Hide the axes
        plt.show()  # Show the image

        batch_1_g[value].append(image_g)  # Add the image to the dictionary
````

### Classification of Test Images

- Classify each image in the test set by finding its nearest neighbour in the training set. For each test image:
  
  1. Calculate the Euclidean (L2) distances of the histogram of the test image and the histogram of every training image. If you have `n` training images, you should calculate `n` distances for every test image.
  
  2. Classify the test image by finding the minimum of the distances. Once you find the minimum distance, the predicted class of the test image is the class of the nearest training image.

  3. Calculate the classification accuracy as the ratio of the number of correctly classified testing images and the total number of testing images.

````python
from scipy.spatial import distance

# training: batch_1_h, test: test_b_h
i= 0
for value in [1,4,8]:
    for image_test in test_b_h[value]:
        distanze = []
        classe = []
        for valuetr in [1,4,8]:
            for image_training in batch_1_h[valuetr]:
                distanze.append(distance.euclidean(image_test, image_training))
                classe.append(valuetr)
        indice = np.where(distanze == min(distanze))[0]
        classe_finale = classe[indice[0]]
        if value == classe_finale:
            i += 1
    
accuracy = i / 30

print('Accuracy finale: ' + str(accuracy))
                

````
