```python
from skimage import io, util, filters
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path = 'woman.png'
image = io.imread(image_path)

plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
```

```python
noisy_image = util.random_noise(image, mode='gaussian', var=0.01) # Gaussian noise with variance 0.01
smoothed_image = filters.gaussian(noisy_image, sigma=1) # Gaussian smoothing with sigma 1


fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].imshow(noisy_image, cmap='gray')
axes[0].set_title('Noisy Image')
axes[0].axis('off')

axes[1].imshow(smoothed_image, cmap='gray')
axes[1].set_title('Smoothed Image')
axes[1].axis('off')
```

```python
# Apply sobel edge detection 
sobel_noisy = filters.sobel(noisy_image)
sobel_smoothed = filters.sobel(smoothed_image)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

axes[0].imshow(sobel_noisy, cmap='gray')
axes[0].set_title('Sobel Edge Detection on Noisy Image')
axes[0].axis('off')

axes[1].imshow(sobel_smoothed, cmap='gray')
axes[1].set_title('Sobel Edge Detection on Smoothed Image')
axes[1].axis('off')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].hist(noisy_image.ravel(), bins=50)
axes[0].set_title('Histogram of Sobel Noisy Image')

axes[1].hist(smoothed_image.ravel(), bins=50)
axes[1].set_title('Histogram of Sobel Smoothed Image')
```

```python
# Threshold values (These might need adjustment)
tn = 0.1  # Example threshold for noisy image
ts = 0.05  # Example threshold for smoothed image
```

```python
# Create binary masks
binary_mask_noisy = sobel_noisy > tn
binary_mask_smoothed = sobel_smoothed > ts
```

```python
# Visualize binary masks
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(binary_mask_noisy, cmap='gray')
ax[0].set_title('Binary Mask - Noisy')
ax[0].axis('off')

ax[1].imshow(binary_mask_smoothed, cmap='gray')
ax[1].set_title('Binary Mask - Smoothed')
ax[1].axis('off')
```
